```python
df = pd.read_csv("survey_2025_data/survey_results_public.csv")
df.head()
```

```python
df.shape
```

```python
print(f"1. Кількість респондентів, які пройшли опитування: {df.shape[0]}")
```

```python
schema = pd.read_csv("survey_2025_data/survey_results_schema.csv")
schema.head()
```

```python
schema.shape
```

```python
question_cols = set(schema["qname"])
valid_cols = list(set(df.columns).intersection(question_cols))
df_questions = df[valid_cols]
complete_responses = df_questions.dropna().shape[0]
print(f"2. Кількість респондентів, які відповіли на всі запитання: {complete_responses}")
```

```python
mean_exp = df["WorkExp"].mean()
median_exp = df["WorkExp"].median()
mode_exp = df["WorkExp"].mode().iloc[0]

print("3. Значення мір центральної тенденції для досвіду (WorkExp) респондентів:")
print(f"Середнє значення досвіду: {mean_exp:.2f} років")
print(f"Медіана досвіду: {median_exp:.2f} років")
print(f"Мода досвіду: {mode_exp:.2f} років")
```

```python
remote_count = df[df["RemoteWork"].isin(["Remote", "Fully remote", "Hybrid"])].shape[0]
print(f"4. Кількість респондентів, які працюють віддалено або гібридно: {remote_count}")
```

```python
python_count = df["LanguageHaveWorkedWith"].str.contains("Python", case=False, na=False).sum()
total_respondents = df["LanguageHaveWorkedWith"].notna().sum()
python_percent = (python_count / total_respondents) * 100

print(f"5. Python використовують {python_count} із {total_respondents} респондентів ({python_percent:.2f}%)")
```

```python
online_count = df["LearnCode"].str.contains("online", case=False, na=False).sum()

print(f"6. Кількість респондентів, які навчалися програмувати через онлайн курси: {online_count}")
```
