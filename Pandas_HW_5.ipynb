###**Завданя 1**
В ноутбуці завантаж таблицю з файлу `applications.csv` в Pandas DataFrame та почисти дані:
- Прибери дублікати applicant_id
- В полі `External Rating` заповни відсутні значення нулями
- В полі `Education level` заповни відсутні значення текстом `Середня`

```python
df_applications = df_applications.drop_duplicates(subset='applicant_id')
df_applications['External Rating'] = df_applications['External Rating'].fillna(0)
df_applications['Education level'] = df_applications['Education level'].fillna('Середня')
```

###**Завданя 2**
Додай до цього DataFrame дані з файлу `industries.csv`, а саме, рейтинги індустрій.

```python
df = df_applications.merge(df_industries, on="Industry", how="left")
```
